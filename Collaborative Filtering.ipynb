```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
movie_path = "movielens/movies.csv"
rate_path = "movielens/ratings.csv"
```

```python
mv_data = pd.read_csv(movie_path)
mv_data.head(5)
```

```python
rt_data = pd.read_csv(rate_path)
rt_data.head(5)
```

```python
mv_data.shape, rt_data.shape
```

```python
rt_data = rt_data.drop('timestamp', axis=1)
```

```python
rt_data.head(5)
```

```python
user_data = pd.merge(rt_data,mv_data,on='movieId')
```

```python
user_data.head(5)
```

```python
user_data.title.unique()
```

```python
mv_to_user=user_data.pivot_table("rating",index='title', columns='userId')
user_to_mv=user_data.pivot_table("rating",index='userId', columns='title')
```

```python
mv_to_user.head(5)
```

```python
user_to_mv.head(5)
```

```python
mv_to_user.fillna(0, inplace=True)
```

```python
mv_to_user.head(5)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
collabor = cosine_similarity(mv_to_user)
collabor
```

```python
collabor = pd.DataFrame(data = collabor, index=mv_to_user.index, columns=mv_to_user.index)
collabor.head(5)
```

```python
def recommand_Search(mv_name):
    return collabor[mv_name].sort_values(ascending=False)[1:6]
```

```python
recommand_Search('"Great Performances" Cats (1998)')
```

```python
recommand_Search("Guardians of the Galaxy (2014)")
```
