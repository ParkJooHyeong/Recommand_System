### 영화 목록, rating을 가지고 영화를 추천해 주기.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
movie_path = "movielens/movies.csv"
rate_path = "movielens/ratings.csv"
```

```python
mv_data = pd.read_csv(movie_path)
rate_data = pd.read_csv(rate_path)
```

```python
mv_data.head(5)
```

```python
rate_data.head(5)
```

```python
mv_data = mv_data.drop("genres", axis=1)
rate_data = rate_data.drop("timestamp", axis=1)
```

```python
data = pd.merge(rate_data, mv_data, on='movieId')
data.head(4)
```

```python
data.shape
```

```python
r_data = data.pivot_table("rating", index="userId", columns="title").fillna(0)
```

```python
r_data.shape
```

```python
r_data.head(4)
```

```python
r_data = r_data.T
```

## SVD(Singular Value Decomposition)

- Q, Sigma, V.T

```python
from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_component=12)
```

```python
decom_mat = SVD.fit_transform(r_data)
```

```python
decom_mat
```

```python
corr = np.corrcoef(decom_mat)
```

```python
corr, corr.shape
```

```python
mv_titles = r_data.T.columns
mv_title_list = list(mv_titles)
```

```python
def recommand_Search(movie_name):
    coeff_t = mv_title_list.index(movie_name)
    coeff_r = corr[coeff_t]
    return list(mv_titles[(coeff_r>=0.9)])[1:20]
```
